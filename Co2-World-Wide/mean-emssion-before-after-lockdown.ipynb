## Import the Dataset and Prepare the Data

```python
import pandas as pd
data = pd.read_csv('./daily-co2-per-contry-2019-2021.csv')
data.head(13)
```

Find distinct contries

```python
areas = data.country.unique()
countries = areas[:13]
print(areas)
print(countries)
top_countries = ["China", "US", "India"]
```

Find distinct sectors

```python
sectors = data.sector.unique()
sectors
```

```python
# convert date to Date object
data['date'] = pd.to_datetime(data['date'])
```

```python
import seaborn as sns
import numpy as np
```

```python
test3 = {"Country": top_countries, "2019 Lockdown Mean": [
], "2020 Lockdown Mean": [], "Difference in Previous and Current Mean (percentage change)": []}

lockdown_start_19 = {"China": "2019-01-23", "US": "2019-03-19", "India": "2019-03-25"}
lockdown_ends_19 = {"China": "2019-03-25",
                    "US": "2019-06-18", "India": "2019-05-31"}
lockdown_start_20 = {"China": "2020-01-23", "US": "2020-03-19", "India": "2020-03-25"}
lockdown_ends_20 = {"China": "2020-03-25", "US": "2020-06-18", "India": "2020-05-31"}

for country in top_countries:
    values_before = data[
        (data['country'] == country)
        & (data['date'] < lockdown_ends_19[country])
        & (data['date'] >= lockdown_start_19[country])
        & ((data['sector'] == "Ground Transport") | (data['sector'] == "Domestic Aviation") | (data['sector'] == "International Aviation"))
    ]['value']
    before = round(values_before.mean(), 2)
    test3["2019 Lockdown Mean"].append(before)

    values_after = data[
        (data['country'] == country)
        & (data['date'] < lockdown_ends_20[country])
        & (data['date'] >= lockdown_start_20[country])
        & ((data['sector'] == "Ground Transport") | (data['sector'] == "Domestic Aviation") | (data['sector'] == "International Aviation"))
    ]['value']
    
    after = round(values_after.mean(), 2)
    test3["2020 Lockdown Mean"].append(after)

    abs_diff = round(after - before, 2)
    per_diff = round(abs_diff / before, 2)
    test3["Difference in Previous and Current Mean (percentage change)"].append(
        "{}({:.0%})". format(abs_diff, per_diff))

test3 = pd.DataFrame(test3)
test3
```

```python
test4 = {"Country": top_countries, "2019 pre-Lockdown Mean": [
], "2020 pre-Lockdown Mean": [], "Difference in Previous and Current Mean (percentage change)": []}

lockdown_start_19 = {"China": "2019-01-23",
                     "US": "2019-03-22", "India": "2019-03-25"}
lockdown_ends_19 = {"China": "2019-03-25",
                    "US": "2019-06-18", "India": "2019-05-31"}
lockdown_start_20 = {"China": "2020-01-23",
                     "US": "2020-03-22", "India": "2020-03-25"}
lockdown_ends_20 = {"China": "2020-03-25",
                    "US": "2020-06-18", "India": "2020-05-31"}

for country in top_countries:
    values_before = data[
        (data['country'] == country)
        & (data['date'] < lockdown_start_19[country])
        & (data['date'] >= "2019-01-01")
        & ((data['sector'] == "Ground Transport") | (data['sector'] == "Domestic Aviation") | (data['sector'] == "International Aviation"))
    ]['value']
    before = round(values_before.mean(), 2)
    test4["2019 pre-Lockdown Mean"].append(before)

    values_after = data[
        (data['country'] == country)
        & (data['date'] < lockdown_start_20[country])
        & (data['date'] >= "2020-01-01")
        & ((data['sector'] == "Ground Transport") | (data['sector'] == "Domestic Aviation") | (data['sector'] == "International Aviation"))
    ]['value']

    after = round(values_after.mean(), 2)
    test4["2020 pre-Lockdown Mean"].append(after)

    abs_diff = round(after - before, 2)
    per_diff = round(abs_diff / before, 2)
    test4["Difference in Previous and Current Mean (percentage change)"].append(
        "{}({:.0%})". format(abs_diff, per_diff))

test4 = pd.DataFrame(test4)
test4
```

```python
for country in top_countries:
    overall_value = data[(data['country'] == country)]['value'].sum()
    for sector in sectors:
        sector_value = data[(data['country'] == country) & (data['sector'] == sector)]['value'].sum()
        percentage = round(sector_value / overall_value, 2)
        print(country, sector, percentage)
```
