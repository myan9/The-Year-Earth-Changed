Calculate the mean of Carbon Emission of Ground Transportaion per conuntry before and after March 1, 2020 (as the covid situaion becomes worse and worse)

```python
import seaborn as sns
import numpy as np
```

India: March 25 2020 - May 31


Comparison:

Before: 01/01/2019 - 01/03/2019
After: 01/03/2019 - 31/12/2019

```python
test1 = {"Country": countries, "2019 Jan-Feb Mean": [
], "2020 Jan-Feb Mean": [], "Difference in Previous and Current Mean (percentage change)": []}

for country in countries:
    values_before = data[
        (data['country'] == country) 
            & (data['date'] < "2019-03-01") 
            & (data['date'] >= "2019-01-01")
            & (data['sector'] == "Ground Transport")
    ]['value']
    before = round(values_before.mean(), 2)
    test1["2019 Jan-Feb Mean"].append(before)

    values_after = data[
        (data['country'] == country)
            & (data['date'] < "2020-03-01") 
            & (data['date'] >= "2020-01-01")
            & (data['sector'] == "Ground Transport")
        ]['value']
    after = round(values_after.mean(), 2)
    test1["2020 Jan-Feb Mean"].append(after)

    abs_diff = round(after - before, 2)
    per_diff = round(abs_diff / before, 2)
    test1["Difference in Previous and Current Mean (percentage change)"].append(
        "{}({:.0%})". format(abs_diff, per_diff))

test1 = pd.DataFrame(test1)
test1
```

```python
test2 = {"Country": countries, "2019 Mar-May Mean": [
], "2020 Mar-May Mean": [], "Difference in Previous and Current Mean (percentage change)": []}

for country in countries:
    values_before = data[
        (data['country'] == country)
        & (data['date'] < "2019-06-01")
        & (data['date'] >= "2019-03-01")
        & (data['sector'] == "Ground Transport")
    ]['value']
    before = round(values_before.mean(), 2)
    test2["2019 Mar-May Mean"].append(before)

    values_after = data[
        (data['country'] == country)
        & (data['date'] < "2020-06-01")
        & (data['date'] >= "2020-03-01")
        & (data['sector'] == "Ground Transport")
    ]['value']
    after = round(values_after.mean(), 2)
    test2["2020 Mar-May Mean"].append(after)

    abs_diff = round(after - before, 2)
    per_diff = round(abs_diff / before, 2)
    test2["Difference in Previous and Current Mean (percentage change)"].append(
        "{}({:.0%})". format(abs_diff, per_diff))

test2 = pd.DataFrame(test2)
test2
```

```python
test3 = {"Country": top_countries, "2019 Lockdown Mean": [
], "2020 Lockdown Mean": [], "Difference in Previous and Current Mean (percentage change)": []}

lockdown_start_19 = {"China": "2019-01-23", "US": "2019-03-19", "India": "2019-03-25"}
lockdown_ends_19 = {"China": "2019-03-25",
                    "US": "2019-06-18", "India": "2019-05-31"}
lockdown_start_20 = {"China": "2020-01-23", "US": "2020-03-19", "India": "2020-03-25"}
lockdown_ends_20 = {"China": "2020-03-25", "US": "2020-06-18", "India": "2020-05-31"}

for country in top_countries:
    values_before = data[
        (data['country'] == country)
        & (data['date'] < lockdown_ends_19[country])
        & (data['date'] >= lockdown_start_19[country])
        & ((data['sector'] == "Ground Transport") | (data['sector'] == "Domestic Aviation") | (data['sector'] == "International Aviation"))
    ]['value']
    before = round(values_before.mean(), 2)
    test3["2019 Lockdown Mean"].append(before)

    values_after = data[
        (data['country'] == country)
        & (data['date'] < lockdown_ends_20[country])
        & (data['date'] >= lockdown_start_20[country])
        & ((data['sector'] == "Ground Transport") | (data['sector'] == "Domestic Aviation") | (data['sector'] == "International Aviation"))
    ]['value']
    
    after = round(values_after.mean(), 2)
    test3["2020 Lockdown Mean"].append(after)

    abs_diff = round(after - before, 2)
    per_diff = round(abs_diff / before, 2)
    test3["Difference in Previous and Current Mean (percentage change)"].append(
        "{}({:.0%})". format(abs_diff, per_diff))

test3 = pd.DataFrame(test3)
test3
```
