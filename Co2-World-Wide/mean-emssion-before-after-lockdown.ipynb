## Some Conlcusions

Before Lockdown: [01/01/2019-01/03/2020)

After Lockdown: [01/03/2020-31/12/2021]

Calculate the mean of Carbon Emission of Ground Transportaion per conuntry before and after March 1, 2020 (as the covid situaion becomes worse and worse)

```python
import seaborn as sns
import numpy as np
```

NOTE: China might need a different time period

Comparison:

Before: 01/01/2019 - 01/03/2019
After: 01/03/2019 - 31/12/2019

```python
test1 = {"Country": countries, "Hisotrical Jan-Feb": [
], "Current Jan-Feb": [], "Difference": []}

for country in countries:
    values_before = data[
        (data['country'] == country) 
            & (data['date'] < "2019-03-01") 
            & (data['date'] >= "2019-01-01")
            & (data['sector'] == "Ground Transport")
    ]['value']
    before = round(values_before.mean(), 2)
    test1["Hisotrical Jan-Feb"].append(before)

    values_after = data[
        (data['country'] == country)
            & (data['date'] < "2020-03-01") 
            & (data['date'] >= "2020-01-01")
            & (data['sector'] == "Ground Transport")
        ]['value']
    after = round(values_after.mean(), 2)
    test1["Current Jan-Feb"].append(after)

    diff = round((after - before) / before, 2)
    test1["Difference"].append("{:.0%}". format(diff))

test1 = pd.DataFrame(test1)
test1
```

```python
test2 = {"Country": countries, "Hisotrical Mar-May": [
], "Current Mar-May": [], "Difference": []}

for country in countries:
    values_before = data[
        (data['country'] == country)
        & (data['date'] < "2019-06-01")
        & (data['date'] >= "2019-03-01")
        & (data['sector'] == "Ground Transport")
    ]['value']
    before = round(values_before.mean(), 2)
    test2["Hisotrical Mar-May"].append(before)

    values_after = data[
        (data['country'] == country)
        & (data['date'] < "2020-06-01")
        & (data['date'] >= "2020-03-01")
        & (data['sector'] == "Ground Transport")
    ]['value']
    after = round(values_after.mean(), 2)
    test2["Current Mar-May"].append(after)

    diff = round((after - before) / before, 2)
    test2["Difference"].append("{:.0%}". format(diff))

test2 = pd.DataFrame(test2)
test2
```
